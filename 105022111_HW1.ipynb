```python
#2
import numpy as np 
import matplotlib.pyplot as plt 
import random   

prob = [0.5, 0.5]   
  
positions = [0] 

rp = np.random.random(100) 
downp = rp < prob[0] 
upp = rp > prob[1] 

for idownp, iupp in zip(downp, upp): 
    down = idownp and positions[-1] > -1
    up = iupp and positions[-1] < 1
    positions.append(positions[-1] - down + up) 
plt.plot(positions) 
plt.show() 
plt.hist(positions)
```
