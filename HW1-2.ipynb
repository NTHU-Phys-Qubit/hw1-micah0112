```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
import math
```

```python
step_n = 5000

path_x = np.random.uniform(-1/2, 1/2, step_n)
path_y = np.random.uniform(-1/2, 1/2, step_n)

path_x = np.cumsum(path_x)
path_y = np.cumsum(path_y)


plt.plot(path_x, path_y)
```

```python
xep = []
yep = []

for i in range(5000):
    step_number = 5000
    
    path_x = np.random.uniform(-1/2, 1/2, step_number)
    path_y = np.random.uniform(-1/2, 1/2, step_number)
    
    xep.append(np.sum(path_x))
    yep.append(np.sum(path_y))
    
plt.plot(xep, yep,'.')
```

```python
ep = []
for i in range(5000):  
    step_number = 5000
    path_x = np.random.uniform(-1/2, 1/2, step_number)
    path_y = np.random.uniform(-1/2, 1/2, step_number)
    
    x = np.sum(path_x)
    y = np.sum(path_y)
    
    ep.append(math.sqrt((x**2)+(y**2)))
    
plt.hist(ep)
```

```python
step_number = 5000
step_set = [-1, 1]

origin = np.zeros((1,2))
step_shape = (step_number,2)

steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])
path = np.concatenate([origin, steps]).cumsum(0)

plt.plot(path[:,0],path[:,1])
```

```python

x_ep = []
y_ep = []

for i in range(5000):
    st_number = 5000
    st_set = [-1, 1]
    origin = np.zeros((1,2))
    
    st_shape = (st_number,2)
    steps = np.random.choice(a=st_set, size=st_shape, p=[0.5, 0.5])
    
    path = np.concatenate([origin, steps]).cumsum(0)
    
    x_ep.append(path[-1][0])
    y_ep.append(path[-1][1])
    
plt.plot(x_ep, y_ep,'.')
```
